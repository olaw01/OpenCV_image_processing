# **OpenCV Morphological Operations**

```python
import cv2
import matplotlib.pyplot as plt
```

```python
Path = "/content/women_pic.jpg"

image = cv2.imread(Path, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```

Erosion

```python
import numpy as np
import cv2 as cv

kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(image,kernel,iterations = 1)

plt.subplot(1,2,1)
plt.imshow(erosion)
plt.title('Erosion')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(image)
plt.title('Original')
plt.axis('off')
plt.show()
```
